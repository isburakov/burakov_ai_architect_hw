5. Заполнение пустых ячеек

5.1 Столбец "Cabin"

```python
#Поскольку число строк с пропущенным значением столбца "Cabin" = 687, необходимо определелить общее количество записей о пассажирах, для принятия решения о целесообразности учета данных значений.
if df.shape[0]- df['Cabin'].isnull().sum()>df['Cabin'].isnull().sum():
  print('Add data')
else:
  print ('Available data =', df.shape[0] - df['Cabin'].isnull().sum(),'.', 'Delete column')
  df.drop(['Cabin'], axis = 1, inplace = True)
df.head(20)
```

Количество пропущенных значений столбца 'Cabin' [687] значительно превышает количество записанных данных [204], вследствии чего принимается решение об удалении столбца 'Cabin'.

5.2 Cтолбец "Embarked"



```python
#Пропущенные значения заполняются наиболее частоповторяющимися значениями
df['Embarked'].value_counts()
```

```python
df['Embarked'].replace(np.nan, 'S', inplace = True)
df.head(20)
```

5.3 Столбец "Age"

```python
#В столбце "Age" вместо пустых ячеек будет вставлено среднее значение возраста отдельновзятых мужчин и женщин.
#Для этого определется количество мужчин и женщин "без возраста".
df.loc[df["Age"].isnull()]["Sex"].value_counts()
```

```python
#Определение среднего возраста мужчин
average_m = (df.loc[df["Sex"] == "male"])["Age"].dropna().mean()
average_m = round(average_m)
print(average_m)
```

```python
#Определение среднего возраста женщин
average_f = (df.loc[df["Sex"] == "female"])["Age"].dropna().mean()
average_f = round(average_f)
print(average_f)
```

Заполнение данных

```python
df["Age"] = df["Age"].fillna(df["Sex"].map({"male":average_m, "female" : average_f}))
```

8. Определение влияния всех признаков на признак Survived

Для выполнения пункта 8 необходимо удалить столбцы не несущие численные значения, исключением является столбец "Sex", в котором необходимо заменить пол человека на 1 - мужской, 0 - женский.

```python
df["Sex_N"] = df["Sex"].map({"male":1, "female":0})
df = df.drop(columns=["Name", "Ticket", "Embarked", "Sex", "PassengerId"])

corr = df.corr()
sns.heatmap(corr,cmap="crest")
corr = df.corr()["Survived"]
corr
```

```python
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier

x = df.drop('Survived', axis=1)
y = df['Survived']

model = RandomForestClassifier()

model.fit(x, y)
importanse = model.feature_importances_

for column, importanse in zip(x.columns, importanse):
    print(column, ':', importanse)
```

Таким образом, больше всего влияет на фактор "Survived":

1. Возраст
2. Стоимость билета
3. Пол

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Меняем кол-во интервалов
bins = 8
# Сопоставляем данные с интервалами
interval = pd.cut(np.array(df["Age"]), bins=np.linspace(0, 80, 9))
iv  = pd.cut(df["Age"], bins, labels=[int(x) for x in range(1, bins + 1)])
df['Age Range'] = iv
df['Range age group'] = interval
df["Survived"] = df["Survived"].map({0:'Deceased', 1:'Survived'})
```

Визуализация результатов исследования выживших в зависимости от возраста

```python
sns.countplot(data=df, x=df['Range age group'], hue = df['Survived'], color = 'r', palette='hls')
sns.set(rc={'figure.figsize':(10,5)})
```

Визуализация результатов исследования выживших в зависимости от стоимости билета

```python
# Меняем кол-во интервалов
bins = 5
# Сопоставляем данные с интервалами
interval2 = pd.cut(np.array(df["Fare"]), bins=np.linspace(0, 520, 5))
iv2  = pd.cut(df["Fare"], bins, labels=[int(x) for x in range(1, bins + 1)])
df['Fare Range'] = iv2
df['Range fare group'] = interval2

sns.countplot(data=df, x=df['Range fare group'], hue = df['Survived'], palette='mako', width = 1)
sns.set(rc={'figure.figsize':(8,8)})
```
